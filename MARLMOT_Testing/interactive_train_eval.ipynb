# **Train Evaluation**

Interactive notebook to evaluate training results

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

Get trained models folder

```python
fpath = r"C:\Users\itber\Documents\learning\self_tutorials\CV_tracking\MARLMOT\trained_models"

logger = pd.read_csv(os.path.join(fpath, "logger.csv"), index_col=0)
metrics = pd.read_csv(os.path.join(fpath, "metrics.csv"), index_col=0)
```

```python
plt.plot(logger.batch_lengths)
plt.title('Batch Lengths')
plt.xlabel('epochs')
plt.ylabel('batch length')
```

```python
plt.plot(logger.avg_return)
plt.title('Undiscounted Average Return')
plt.xlabel('epochs')
plt.ylabel('return')
```

```python
plt.plot(metrics.approx_kl)
plt.title("Approximate KL Divergence")
plt.xlabel("epochs")
plt.ylabel("KL Divergence");
```

```python
plt.plot(metrics.false_positives, label='false_positives')
plt.plot(metrics.false_negatives, label='false_negatives')
plt.plot(metrics.mismatch_errrors, label='mismatch_errrors')
plt.plot(metrics.cost_penalty, label='cost_penalty')
plt.title("Accuracy Metrics")
plt.xlabel("Epochs")
plt.ylabel("Number")
plt.legend();
```

```python
plt.plot(metrics.action_0_ratio, label='0')
plt.plot(metrics.action_1_ratio, label='1')
plt.plot(metrics.action_2_ratio, label='2')
plt.plot(metrics.action_3_ratio, label='3')
plt.plot(metrics.action_4_ratio, label='4')
plt.title("Action Ratios")
plt.xlabel("Epochs")
plt.ylabel("Ratio")
plt.legend();
```
